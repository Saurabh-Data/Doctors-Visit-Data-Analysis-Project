```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('DoctorVisits-DA.csv')
data.head()  # Preview the first few rows
```

```python
duplicates = data.duplicated().sum()
print(f"Number of duplicate entries: {duplicates}")
```

```python
sns.countplot(x='gender', hue='illness', data=data)
plt.title('Gender-wise Illness Distribution')
plt.show()
```

```python
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title('Missing Values Heatmap')
plt.show()
```

```python
numeric_data = data.select_dtypes(include=[float, int])

# Calculate the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Between Variables')
plt.show()
```

```python
# Check the column names to confirm
print(data.columns)

# Plotting the scatter plot with the correct column names
plt.figure(figsize=(8, 6))
sns.scatterplot(x='income', y='visits', data=data)
plt.title('Income vs Hospital Visits')
plt.xlabel('Income')
plt.ylabel('Hospital Visits')

plt.savefig('/content/Income_vs_Hospital_Visits.png')
plt.show()
```
